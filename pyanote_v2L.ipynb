```python
import os

# Definición de la ruta del directorio a examinar
dir_path = '/content/drive/MyDrive/Diarization/mickey/'

# Inicialización de un diccionario vacío que almacenará los nombres de los archivos de audio
audio_files = {}

# Recorrer todos los archivos en el directorio
for i, file in enumerate(os.listdir(dir_path)):
    # Verificar si el nombre del archivo comienza con "Muestra"
    if file.startswith('Muestra'):
        # Si es así, agregarlo al diccionario con una clave tipo 'audio1', 'audio2', etc.
        audio_files[f'audio{i+1}'] = file

# Imprimir los archivos recopilados
for audio_var, audio_file in audio_files.items():
    print(f'{audio_var}: {audio_file}')



```

**Ahora realizamos la dyarizacion en si**

```python
for audio_var, audio_file in audio_files.items():
    print(f'{audio_var}: {audio_file}')
    print("se esta realizando la diarization de "+audio_file)
    diarization = pipeline('/content/drive/MyDrive/Diarization/mickey/'+audio_file)
    print("primer archivo diarizado")
    print("Guardando rttm de "+audio_var)
    with open(f"/content/drive/MyDrive/Diarization/rttmmickey/{audio_var}.rttm", "w") as rttm:
      diarization.write_rttm(rttm)
```

```python
# Definición de la ruta del archivo a leer
file_path = '/content/drive/MyDrive/Diarization/rttmmickey/audio2.rttm'

# Abrir el archivo en modo de lectura
with open(file_path, 'r') as rttm_file:
    # Leer el contenido completo del archivo
    rttm_content = rttm_file.read()

# Imprimir el contenido del archivo
print(rttm_content)
```

**Ahora con el audio ya dyarizado en un fichero de texto, vamos a cortar el audio**

```python
import pandas as pd

# Asumiendo que `rttm_content` contiene el contenido del archivo RTTM como una cadena de texto
raw_data = rttm_content

# Divide el contenido en líneas individuales y elimina espacios en blanco
lines = [line.strip() for line in raw_data.strip().split('\n') if line.strip()]

# Lista para almacenar la información extraída de cada línea
data = []

# Procesa cada línea
for line in lines:
    # Divide la línea en partes individuales (los archivos RTTM tienen un formato espaciado)
    parts = line.split()

    # Extrae la información relevante y la almacena en un diccionario
    speaker_info = {
        'Audio': parts[1],          # Nombre del archivo de audio
        'Channel': int(parts[2]),   # Canal (por lo general, 1 para mono, 1 o 2 para estéreo)
        'Start_time': float(parts[3]), # Tiempo de inicio del segmento de habla
        'Duration': float(parts[4]),   # Duración del segmento de habla
        'Speaker': parts[7]         # ID del hablante
    }

    # Agrega el diccionario a la lista de datos
    data.append(speaker_info)

```

```python
df = pd.DataFrame(data)
```

**Ahoora cortamos**

```python
# Lista para almacenar los segmentos cortados
cut_audio_segments = []

# Realiza los cortes en el archivo de audio utilizando la información del DataFrame
for index, row in filtered_df.iterrows():
    start_time = row['Start_time'] * 1000  # Convierte a milisegundos
    end_time = (row['Start_time'] + row['Duration']) * 1000  # Convierte a milisegundos
    cut_segment = audiofirst[start_time:end_time]
    cut_audio_segments.append(cut_segment)
```

**Unimos los segmentos cortados**

```python
concatenated_audio = AudioSegment.empty()

for segment in cut_audio_segments:
    concatenated_audio += segment
```
